```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import os
import csv
import numpy as np
import matplotlib.pyplot as plt

# Read frame counts from CSV file
with open('/data/jong980812/project/VideoMAE_experiments/frame_counts_video.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # skip header row
    frame_counts = [int(row[1]) for row in reader]

# Calculate counts in each 30-frame interval
bin_counts, bin_edges = np.histogram(frame_counts, bins=np.arange(0, max(frame_counts)+10, 10),range=(0,100))

# Plot histogram
plt.bar(bin_edges[:-1], bin_counts, width=10)
plt.xlabel('Frame')
plt.ylabel('Count')
plt.xlim(0, 100)  # set X-axis limits

plt.xticks(np.arange(0, 101, 10))


plt.show()


# Read the original CSV file into a pandas DataFrame
df = pd.read_csv("/data/jong980812/project/VideoMAE_experiments/dataset/epic/epic100_val.csv", header=None)

# Drop the second column
df.drop(columns=[1], inplace=True)

# Add ".mp4" to the end of the first column
df[0] = df[0].apply(lambda x: x+".mp4")

# Rename the columns
# df.columns = ['Video file', 'Frame count']
print(df)
# Save the new CSV file
df.to_csv("/data/jong980812/project/VideoMAE_experiments/dataset/epic/epic100_noun_val.csv", index=False, sep=' ',header=None)
# 
```
